```python
import pandas as pd
import numpy as np
from datetime import date
```

```python
'''
    Before reading the csv, some work was done on the file:
    - drop columns name and region (we can id the district with the code provided)
    - drop column no. of cities (useless since we have the ratio of urban habitatnts)
    - Rename unemploymant rate '95 and '96 to unemployment_95 and unemployment_96, respectively
    - Rename no. of commited crimes '95 and '96 to crimes_95 and crimes_96, respectively

'''

demographics = pd.read_csv("district_updated.csv", sep=",", na_values=["?"])
```

**Calculate Crimes per 1k instead of all crimes**

```python
def normalize_crime(row, year):
    if year == 1995:
        return int(row['crimes_95'])/row['no. of inhabitants'] * 1000
    elif year == 1996:
        return row['crimes_96']/row['no. of inhabitants'] * 1000
```

```python
demographics['unemployment_95'].fillna(demographics['unemployment_96'], inplace=True)
demographics['unemployment_96'].fillna(demographics['unemployment_95'], inplace=True)
demographics['crimes_95'].fillna(demographics['crimes_96'], inplace=True)
demographics['crimes_96'].fillna(demographics['crimes_95'], inplace=True)


#demographics['crimes_95_per_1k'] = demographics.apply(lambda row: normalize_crime(row, 1995), axis=1)
```

```python
demographics['crimes_95'] = demographics['crimes_95'].astype(int)
demographics.dtypes
```

```python
demographics['crimes_95_per_1k'] = demographics.apply(lambda row: normalize_crime(row, 1995), axis=1)
```

```python
demographics['crimes_96_per_1k'] = demographics.apply(lambda row: normalize_crime(row, 1996), axis=1)
```

```python
demographics.head()
```

```python
demographics = demographics.drop(demographics.columns[[2, 3, 4, 5, 10]], axis = 1)
```

```python
demographics.head()
```

```python
def increase(row, type):
    if type == 'unemployment':
        return row['unemployment_96'] - row['unemployment_95']

    elif type == 'crime':
        return row['crimes_96_per_1k'] - row['crimes_95_per_1k']
```

```python
def average(row, type):
    if type == 'unemployment':
        return (row['unemployment_96'] + row['unemployment_95'] )/ 2

    elif type == 'crime':
        return (row['crimes_96_per_1k'] + row['crimes_95_per_1k']) / 2
```

```python
demographics['crime_increase'] = demographics.apply(lambda row : increase(row, "crime"), axis=1)
```

```python
demographics['unemployment_increase'] = demographics.apply(lambda row : increase(row, "unemployment"), axis=1)
```

```python
demographics['crime_avg'] = demographics.apply(lambda row : average(row, "crime"), axis=1)
```

```python
demographics['unemployment_avg'] = demographics.apply(lambda row : average(row, "unemployment"), axis=1)
```

```python
demographics = demographics.drop(demographics.columns[[4, 5, 6, 7, 8, 9]], axis = 1)
```
